```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
chat = '/content/drive/MyDrive/WhatsApp Chat with Kishan Bhai Personal.txt'

# importing the text file
with open(chat, 'r', encoding='utf-8') as file:
    chat_data = file.readlines()

# Join all lines into a single string
chat = ' '.join(chat_data)

print(chat)
```

```python
import nltk
nltk.download('punkt')

import nltk
nltk.download('stopwords')

def clean_text(text):
    # Remove unnecessary metadata (timestamps, etc.)
    text = re.sub(r'\d{1,2}/\d{1,2}/\d{2,4}, \d{1,2}:\d{2} [APM]{2} - ', '', text)

    # Remove special characters
    text = re.sub(r'[^\w\s]', '', text)

    # Convert to lowercase
    text = text.lower()

    return text
```

```python
# Clean the chat text
cleaned_text = clean_text(chat)

# Tokenization
tokens = word_tokenize(cleaned_text)

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

# Initialize stemming
porter = PorterStemmer()
stemmed_tokens = [porter.stem(word) for word in filtered_tokens]

# Join tokens back into a single string
processed_text = ' '.join(stemmed_tokens)

print(processed_text)
```

```python
#Step 4- Creating Word Clouds of the processed chat data

# Generate WordCloud
wordcloud = WordCloud(width=800, height=250, background_color='grey').generate(processed_text)

# Display the WordCloud
plt.figure(figsize=(20, 15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('WhatsApp Chat WordCloud')
plt.show()


 #Specify the path where you want to save
save_path = r'C:\Users\Asus\Dropbox\My PC (LAPTOP-Q5SI23RP)\Downloads\wordcloud.png'  # Change this path to your desired location

# Save WordCloud as image
wordcloud.to_file(save_path)

print(f"Word Cloud saved as image: {save_path}")
```

```python
#Step 5- Extracting information through sentiment analysis

from textblob import TextBlob

# Function to perform sentiment analysis using TextBlob
def sentiment_analysis(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

# Perform sentiment analysis on processed_text
sentiment_score = sentiment_analysis(processed_text)

if sentiment_score > 0:
    sentiment_label = "Positive"
elif sentiment_score < 0:
    sentiment_label = "Negative"
else:
    sentiment_label = "Neutral"

print(f"Sentiment Analysis Result: {sentiment_label} (Score: {sentiment_score})")



from textblob import TextBlob

# Function to perform sentiment analysis using TextBlob
def sentiment_analysis(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

# Perform sentiment analysis on processed_text
sentiment_score = sentiment_analysis(processed_text)

if sentiment_score > 0:
    sentiment_label = "Positive"
elif sentiment_score < 0:
    sentiment_label = "Negative"
else:
    sentiment_label = "Neutral"

print(f"Sentiment Analysis Result: {sentiment_label} (Score: {sentiment_score})")
```
